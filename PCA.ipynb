# Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(15,15))
dendogram=sch.dendrogram(sch.linkage(df1,method='complete', metric='euclidean',))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ag=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='complete')
```

```python
cluster_id=ag.fit_predict(df1_values)
```

```python
cluster_id=pd.DataFrame(cluster_id,columns=['Clusters'])
```

```python
df1['cluster_id']=ag.labels_
```

```python
df1
```

```python
df1.cluster_id.value_counts()
```

# KMeans

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,14):
    kmeans=KMeans(n_clusters=i,random_state=42)
    kmeans.fit(df1_values)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,14),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
cluster_new=KMeans(3,random_state=42)
cluster_new.fit(df1)
```
